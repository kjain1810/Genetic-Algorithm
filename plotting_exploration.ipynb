```python
import json
import math
```

```python
with open('exploring.json') as f:
    res = json.load(f)
```

```python
val_dict = {}
val_dict = {"0": [], "1": [], "2": [], "3": [], "4": [], "5": [], "6":[]}
for i in res["explorations"]:
    val_dict[str(i["index"])].append({"difference":i["difference"], "results": i["results"]})
```

```python
print(val_dict)
```

```python
def fitness(res, FACTOR=1.3):
    return res[0] * FACTOR + res[1]
orig_res = [79589592533.03973, 90480460590.95456]
```

```python
indices = [0, 1 ,2, 3 , 4, 5 , 6]
for i in indices:
    x = []
    y = []
    for j in val_dict[str(i)]:
        x.append(math.log(abs(j["difference"]))*(abs(j["difference"])/j["difference"]))
        y.append((fitness(j["results"]) - fitness(orig_res))/1e11)
        
    plt.plot(x, y, 'o')
    plt.xlabel(i)
    plt.show()
```
